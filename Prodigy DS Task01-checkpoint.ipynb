## How the country population change over time.

```python
columns = ['2001', '2002', '2003', '2004', '2005', '2006', '2007',
        '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016',
        '2017', '2018', '2019', '2020', '2021', '2022']
for years in columns:
    fig = plt.figure(figsize=(5, 5))
    plt.hist(df[years], color='magenta', bins=15)
    plt.title(f'Population of {years} Over Years')
    plt.xlabel('range of Year')
    plt.ylabel('Population count')
    plt.grid(True)
    plt.show()
```

## Display top 10 Population in 2001 & how its changes over the years

```python
top_10_Population_2001 = df.sort_values(by='2001', ascending=False).head(10)
print(top_10_Population_2001)
```

```python
# Transpose DataFrame for plotting
top_10_population_by_2001_t = top_10_Population_2001.set_index('Country Name').T
```

```python
for country_name, data_values in top_10_population_by_2001_t.iterrows():
    fig = plt.figure(figsize=(8, 5))
    sns.barplot(x=data_values.index, y=data_values.values)
    plt.xlabel('population by sectors')
    plt.ylabel('Population')
    plt.title(f"{country_name}")
    plt.xticks(rotation=90)
    plt.show()
```

## Display Smallest 10 countries in 2001 & how its population changes over the years

```python
smallest_10_country_2001 = df.sort_values(by='2001').head(10)
print(smallest_10_country_2001)
```

```python
# Transpose DataFrame for plotting
smallest_10_country_by_2001_t = smallest_10_country_2001.set_index('Country Name').T
```

```python
for country_name, data_values in smallest_10_country_by_2001_t.iterrows():
    fig = plt.figure(figsize=(15, 10))
    sns.barplot(x=data_values.index, y=data_values.values)
    plt.xlabel('Countries')
    plt.ylabel('Population')
    plt.title(f"{country_name}")
    plt.xticks(rotation=90)
    plt.show()
```
